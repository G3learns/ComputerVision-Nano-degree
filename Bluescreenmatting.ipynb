```python
import cv2
from matplotlib import image as im
from matplotlib import pyplot as plt
```

```python
img=im.imread("/content/drive/MyDrive/Colab Notebooks/Image/pizza_bluescreen.jpg")
```

```python
import numpy as np
```

```python
lowthres=np.array([5,5,200])
highthres=np.array([245,245,255])
mask=cv2.inRange(img,lowthres,highthres)
plt.imshow(mask,cmap='gray')
```

```python
imagew=np.copy(img)
```

```python
imagew[mask!=0]=[0,0,0]
```

```python
plt.imshow(imagew)
imagew.shape
```

```python
img1=im.imread("/content/drive/MyDrive/Colab Notebooks/Image/R.jfif")
imgw1=np.copy(img1)
imgw1=imgw1[0:514,0:816,0:3]
imgw1[mask==0]=[0,0,0]
plt.imshow(imgw1)
```

```python
img2=imagew+imgw1
plt.imshow(img2)
```
